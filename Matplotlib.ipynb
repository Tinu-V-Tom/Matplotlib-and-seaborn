```python
import matplotlib.pyplot as plt
```

```python
#plot() is used to plot the points
#you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.
#you can use the title() function to set a title for the plot.
#you can use the loc argument in title() function to set the location of title for the plot.
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x,y)
plt.title("Sample Diagram ",loc='left')
plt.xlabel("X points")
plt.ylabel("Y points")
plt.show()
```

```python
#to plot only the markers, you can use shortcut string notation parameter.

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x,y,'*')
plt.title("Sample Diagram ")
plt.xlabel("X points")
plt.ylabel("Y points")
plt.show()
```

```python
#to plot only the markers along with line plot, you can use shortcut string notation parameter.

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x,y,marker='P',linestyle='--',color='r')
plt.title("Sample Diagram ")
plt.xlabel("X points")
plt.ylabel("Y points")
plt.show()
```

```python
#You can use the keyword argument markersize or the shorter version, ms to set the size of the markers:
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x,y,marker='o',linestyle='--',color='g',ms=20)
plt.title("Sample Diagram ")
plt.xlabel("X points")
plt.ylabel("Y points")
plt.show()
```

```python
#You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:
#You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers
#You can use the keyword argument linewidth or the shorter lw to change the width of the line.
#You can use the keyword argument linestyle or the shorter ls to change the style of the line.
#
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x,y,marker='o',ls='--',lw=7,color='k',ms=15,mec='k',mfc='r')
plt.title("Sample Diagram ")
plt.xlabel("X points")
plt.ylabel("Y points")
plt.show()
```

```python
#You can plot as many lines as you like by simply adding more plt.plot() functions:
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x)
plt.plot(y)
plt.title("Sample Diagram ")
plt.xlabel("X points")
plt.ylabel("Y points")
plt.show()
```

```python
#you can use the grid() function to add grid lines to the plot
#You can use the axis parameter in the grid() function to specify which grid lines to display. By default the value is both.
#You can also set the line properties of the grid, like this: grid(color = 'color', linestyle = 'linestyle', linewidth = number
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x)
plt.plot(y)
plt.title("Sample Diagram ")
plt.xlabel("X points")
plt.ylabel("Y points")
plt.grid(axis='y')
plt.show()
```

```python
#With the subplot() function you can draw multiple plots in one figure
#The subplot() function takes three arguments that describes the layout of the figure.
#The layout is organized in rows and columns, which are represented by the first and second argument.
#The third argument represents the index of the current plot.
#You can add a title to the entire figure with the suptitle() function which stands for supertitle


import numpy as np
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 6])
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("sample figure 1 ")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 40, 30, 20])
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Sample figure 2")

plt.suptitle("Sample Subplot Diagram")

plt.show()
```

```python
#you can use the scatter() function to draw a scatter plot.
#The scatter() function plots one dot for each observation.
#It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.scatter(x,y)
plt.title("Sample Diagram ")
plt.xlabel("X points")
plt.ylabel("Y points")
plt.show()
```

```python
#You can adjust the transparency of the dots with the alpha argument.
# You can change the size of the points using s
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
sizes = [20,50,100,200,500,1000,60,90,10,300,600,800,75]
plt.scatter(x, y, s=sizes, alpha=0.5)
plt.show()
```

```python
# you can use the bar() function to draw bar graphs:
#The bar() function takes arguments that describes the layout of the bars.
#The categories and their values represented by the first and second argument as arrays.
x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]
plt.bar(x,y)
plt.show()
```

```python
#If you want the bars to be displayed horizontally instead of vertically, use the barh() function:
x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]
plt.barh(x,y)
plt.show()
```

```python
#take the keyword argument color to set the color of the bars:
#he bar() takes the keyword argument width to set the width of the bars,For horizontal bars, use height instead of width.
x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]
plt.bar(x,y,color='g',width=0.2)
plt.show()
```

```python
#A histogram is a graph showing frequency distributions.
#we use the hist() function to create histograms.
#The hist() function will use an array of numbers to create a histogram, the array is sent into the function as an argument.
import numpy as np
x = np.random.normal(0, 1, 250)
plt.hist(x,color='hotpink')
plt.show() 
```

```python
#you can use the pie() function to draw pie charts
y = [35, 25, 25, 15]
plt.pie(y)
plt.show() 
```

```python
#Add labels to the pie chart with the labels parameter
# the explode parameter of the pyplot can be used to move the wedges stand out.
#If you need to add shadow the plot, set shadow=True in the pie()
#If you need to add colors to each wedge,that can be done by passing colors array
#To add explanation of each wedge by using legend()
#To add a header to the legend, add the title parameter to the legend functio
y = np.array([35, 25, 25, 15])
mylabels = ["A", "B", "C", "D"]
explode=[0.2,0,0,0]
c=['k','r','g','y']
plt.pie(y, labels = mylabels,explode=explode,shadow=True,colors=c)
plt.legend(title='Four letters')
plt.show()
```
