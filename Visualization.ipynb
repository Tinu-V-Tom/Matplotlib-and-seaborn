```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic dataset
df = pd.read_csv('titanic.csv') 
```

```python
plt.hist(df['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
survived_counts = df['Survived'].value_counts()
plt.bar(['Not Survived', 'Survived'], survived_counts, color=['red', 'green'])
plt.title('Survival Count')
plt.ylabel('Number of People')
plt.show()
```

```python
pclass_counts = df['Pclass'].value_counts()
plt.pie(pclass_counts, labels=['3rd', '1st', '2nd'], autopct='%1.1f%%', startangle=90)
plt.title('Passenger Class Distribution')
plt.axis('equal')
plt.show()
```

```python
survived_0 = df[df['Survived'] == 0]['Age'].dropna()
survived_1 = df[df['Survived'] == 1]['Age'].dropna()
plt.boxplot([survived_0, survived_1], labels=['Not Survived', 'Survived'])
plt.title('Age Distribution by Survival')
plt.ylabel('Age')
plt.show()
```

```python
plt.scatter(df['Age'], df['Fare'], alpha=0.5, c=df['Survived'], cmap='coolwarm')
plt.title('Fare vs Age (Colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.colorbar(label='Survived')
plt.show()
```

```python
cumulative = df['PassengerId'].sort_values()
plt.plot(cumulative)
plt.title('Cumulative Passenger Count')
plt.xlabel('Index')
plt.ylabel('Passenger ID')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Optional: to use seaborn's built-in styling
sns.set(style="whitegrid")
```

```python
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title('Survival Count')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived', palette='Set1')
plt.suptitle('Pairplot of Numeric Features Colored by Survival', y=1.02)
plt.show()
```

```python
correlation = df[['Age', 'Fare', 'Pclass', 'SibSp', 'Parch', 'Survived']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.violinplot(x='Survived', y='Age', data=df, palette='muted', split=True)
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.title('Violin Plot of Age by Survival')
plt.show()
```
